```python
import pickle 
import pandas as pd
import numpy as np

with open('./decision_model.pkl', 'rb') as f:
    decision_model = pickle.load(f)

base = pd.read_csv("./random_selected_test_rows.csv")
del base[base.columns[0]]
base.head()
```

```python
base['class'].unique()
```

```python
classes = ['extension', 'finger_abduction', 'finger_adduction', 'flexion', 'grip',
 'pronation', 'radial_deviation', 'rest', 'supination', 'ulnar_deviation']
decision_model
```

```python
target_cols = 'class'
train_cols = ['SENSOR0_RMS', 'SENSOR1_RMS', 'SENSOR2_RMS', 'SENSOR3_RMS', 'SENSOR0_APPEN', 'SENSOR1_APPEN', 'SENSOR2_APPEN', 'SENSOR3_APPEN']
```

```python
filtered_base = base.loc[(base['class'] == "grip") | (base['class'] == "rest") | (base['class'] == "flexion")]
filtered_base['class'].value_counts()
```

```python
def select_random_data():
    r_row = np.random.randint(0, len(filtered_base))
    test_one = filtered_base.iloc[r_row]

    X = test_one[train_cols]
    Y = test_one[target_cols]
    X, Y

    predict_classes = decision_model.predict([X])
    pred_class = classes[predict_classes[0]].upper()
    print("Classe Original: ", Y.upper())
    print("Classe Predita: ", pred_class)
    return pred_class
```

```python
import connection
connection.con()
```

```python
random_pred = select_random_data()
```
