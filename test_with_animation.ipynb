```python
import pickle 
import pandas as pd
import numpy as np

with open('./knn_model.pkl', 'rb') as f:
    knn_model = pickle.load(f)

base = pd.read_csv("./random_selected_test_rows.csv")
del base[base.columns[0]]
base.head()
```

```python
base['class'].unique()
```

```python
classes = ['extension', 'finger_abduction', 'finger_adduction', 'flexion', 'grip',
 'pronation', 'radial_deviation', 'rest', 'supination', 'ulnar_deviation']
knn_model
```

```python
target_cols = 'class'
train_cols = ['SENSOR0_RMS', 'SENSOR1_RMS', 'SENSOR2_RMS', 'SENSOR3_RMS', 'SENSOR0_APPEN', 'SENSOR1_APPEN', 'SENSOR2_APPEN', 'SENSOR3_APPEN']
```

```python
t = base.loc[(base['class'] == "grip") | (base['class'] == "rest") | (base['class'] == "flexion")]
t['class'].value_counts()
```

```python
# TODO: Temporary uses only GRIP, REST and FLEXION
t_base = base.loc[(base['class'] == "grip") | (base['class'] == "rest") | (base['class'] == "flexion")]

r_row = np.random.randint(0, len(t_base))
test_one = t_base.iloc[r_row]

X = test_one[train_cols]
Y = test_one[target_cols]
X, Y
```

```python
import connection
connection.con()
```

```python
predict_classes = knn_model.predict([X])
classes[predict_classes[0]]
```

```python
connection.send_movement(classes[predict_classes[0]].upper())
```
